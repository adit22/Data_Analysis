```python
new_cols=["Daily_Confirmed_Ratio","Total_Confirmed_Ratio","Daily_Recovered_Ratio","Total_Recovered_Ratio","Daily_Deceased_Ratio","Total_Deceased_Ratio"]
existing_cols=list(datajly_ratio_analysis.columns)
existing_cols.pop(0)
existing_cols
prv_vals=[18256,585795,12565,347840,506,17411]
for i,j,l in zip(new_cols,existing_cols,prv_vals):
    datajly_ratio_analysis[i]=list(range(len(datajly_ratio_analysis)))
    datajly_ratio_analysis[i][0]=datajly_ratio_analysis[j][0]/l
    tmp=[]
    tmp.append(datajly_ratio_analysis[i][0])
    for k in range(0,len(datajly_ratio_analysis)-1):
        tmp.append(datajly_ratio_analysis[j][k+1]/datajly_ratio_analysis[j][k])
    datajly_ratio_analysis[i]=tmp
datajly_ratio_analysis.head()
        
```

```python
dcr=datajly_ratio_analysis.iloc[-7:,7:8]
dcr.reset_index(inplace = True, drop = True)
dc=datajly_ratio_analysis.iloc[-7:,1:2]
dc.reset_index(inplace = True, drop = True)
```

```python
next_day_confirmed_case_count_mean_based=dcr.mean()[0]*dc.mean()[0]
next_day_confirmed_case_count_mean_based
```

```python
next_day_confirmed_case_count_max_based=max(dcr["Daily_Confirmed_Ratio"])*max(dc["Daily Confirmed"])
next_day_confirmed_case_count_max_based
```

```python
next_day_confirmed_case_count_mean_max_based=dcr.mean()[0]*max(dc["Daily Confirmed"])
next_day_confirmed_case_count_mean_max_based
```

```python
next_day_confirmed_case_count_max_mean_based=max(dcr["Daily_Confirmed_Ratio"])*dc.mean()[0]
next_day_confirmed_case_count_max_mean_based
```

```python
next_day_confirmed_case_count_mean_last_based=dcr.mean()[0]*dc["Daily Confirmed"][len(dc)-1]
next_day_confirmed_case_count_mean_last_based
```

```python
next_day_confirmed_case_count_last_mean_based=dcr["Daily_Confirmed_Ratio"][len(dcr)-1]*dc.mean()[0]
next_day_confirmed_case_count_last_mean_based
```

```python
next_day_confirmed_case_count_last_based=dcr["Daily_Confirmed_Ratio"][len(dcr)-1]*dc["Daily Confirmed"][len(dc)-1]
next_day_confirmed_case_count_last_based
```

```python
next_day_confirmed_case_count_middle_based=dcr["Daily_Confirmed_Ratio"][3]*dc["Daily Confirmed"][3]
next_day_confirmed_case_count_middle_based
```

```python
pred_val=[]
pred_val.append(np.round(next_day_confirmed_case_count_mean_based))
pred_val.append(np.round(next_day_confirmed_case_count_max_based))
pred_val.append(np.round(next_day_confirmed_case_count_mean_max_based))
pred_val.append(np.round(next_day_confirmed_case_count_max_mean_based))
pred_val.append(np.round(next_day_confirmed_case_count_mean_last_based))
pred_val.append(np.round(next_day_confirmed_case_count_last_mean_based))
pred_val.append(np.round(next_day_confirmed_case_count_last_based))
pred_val.append(np.round(next_day_confirmed_case_count_middle_based))
```

```python
pred_dict={"date":datajly_ratio_analysis["Date"][len(datajly_ratio_analysis)-1]+datetime.timedelta(hours=24),"original_val":0,"mean_based":pred_val[0],"max_based":pred_val[1],"mean_max_based":pred_val[2],"max_mean_based":pred_val[3],"mean_last_based":pred_val[4],"last_mean_based":pred_val[5],"last_based":pred_val[6],"middle_based":pred_val[7]}
```
