```python
edf=dce.iloc[0:len(dce)-1,10:]
edf
```

```python

for j,k in zip(range(len(edf)),range(len(dce)-1)):
    
    a=list(edf.loc[j])
    pos=[]
    neg=[]
    for i in a:
        if(i>=0):
            pos.append(i)
        else:
            neg.append(i)
    ps=min(pos)
    ng=max(neg)
    psi=a.index(ps)
    ngi=a.index(ng)
    psc=list(edf.columns)[psi]
    ngc=list(edf.columns)[ngi]
    
    
    print("for date: "+str(dce["date"][k]).strip("00:00:00")+" the lowest positve error was by the method: "+psc+" and the highest negative error by the method: "+ngc)
```

```python
err_dict={"date":[],"mipe":[],"mxne":[],"mipv":[],"mxnv":[],"original_val":[],"pprd_val":[],"nprd_val":[]}
psc=[]
ngc=[]
date=[]
ps=[]
ng=[]
org_val=[]
pprd_val=[]
nprd_val=[]
for j,k in zip(range(len(edf)),range(len(dce)-1)):
    
    a=list(edf.loc[j])
    pos=[]
    neg=[]
    for i in a:
        if(i>=0):
            pos.append(i)
        else:
            neg.append(i)
    aa=min(pos)
    bb=max(neg)
    psi=a.index(aa)
    ngi=a.index(bb)
    ps.append(aa)
    ng.append(bb)
    psc.append(list(edf.columns)[psi])
    ngc.append(list(edf.columns)[ngi])
    date.append(str(dce["date"][k]).strip("00:00:00"))
    org_val.append(dce["original_val"][k])
    pprd_val.append(dce["original_val"][k]-aa)
    nprd_val.append(dce["original_val"][k]-bb)
    
err_dict.update({"date":date})
err_dict.update({"mipe":psc})
err_dict.update({"mxne":ngc})
err_dict.update({"mipv":ps})
err_dict.update({"mxnv":ng})
err_dict.update({"original_val":org_val})
err_dict.update({"pprd_val":pprd_val})
err_dict.update({"nprd_val":nprd_val})


```
