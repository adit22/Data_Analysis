```python
#finding min_max of numeric columns
```

```python
min_max_dict={}
for i in num_col:
    min_max_dict.update({i:[min(data[i]),max(data[i])]})
    
    
min_max_dict
```

```python
min_max_dict.values()
```

```python
#saving mean and standard deviation from original dataset
data_mean_std=[]
for i in list(num_col):
    data_mean_std.append([i,np.mean(data[i]),np.std(data[i])])
    
    
df_mean_std = pd.DataFrame(data_mean_std, columns = ['col_name', 'mean','std_dev'])
#df_mean_std
```
