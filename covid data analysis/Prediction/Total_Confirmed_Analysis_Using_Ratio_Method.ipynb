```python
data_total_ratio=data_total.copy()
```

```python
new_cols=["Daily_Confirmed_Ratio","Total_Confirmed_Ratio","Daily_Recovered_Ratio","Total_Recovered_Ratio","Daily_Deceased_Ratio","Total_Deceased_Ratio"]
existing_cols=list(data_total_ratio.columns)
existing_cols.pop(0)
existing_cols
for i,j in zip(new_cols,existing_cols):
    data_total_ratio[i]=list(range(len(data_total_ratio)))
    data_total_ratio[i][0]=1
    tmp=[]
    tmp.append(data_total_ratio[i][0])
    for k in range(0,len(data_total_ratio)-1):
        tmp.append(data_total_ratio[j][k+1]/data_total_ratio[j][k])
    data_total_ratio[i]=tmp
data_total_ratio.head()
        
```

```python
strt_cnt=1
end_cnt=8
strt_lst=[]
end_lst=[]
strt_lst.append(strt_cnt)
end_lst.append(end_cnt)
while(end_cnt<=len(data_total_ratio)):
        strt_cnt+=1
        strt_lst.append(strt_cnt)
        end_cnt+=1
        end_lst.append(end_cnt)
if(end_lst[len(end_lst)-1]>=len(data_total_ratio)):
    end_lst.pop(len(end_lst)-1)
strt_lst.pop(len(strt_lst)-1)
#print(strt_lst)
#print(end_lst)
#print(len(strt_lst)==len(end_lst))

for st,en in zip(strt_lst,end_lst):

    tcr=data_total_ratio.iloc[st:en,8:9]
    tcr.reset_index(inplace = True, drop = True)
    tc=data_total_ratio.iloc[st:en,2:3]
    tc.reset_index(inplace = True, drop = True)



    next_day_tconfirmed_case_count_mean_based=tcr.mean()[0]*tc.mean()[0]

    next_day_tconfirmed_case_count_max_based=max(tcr["Total_Confirmed_Ratio"])*max(tc["Total Confirmed"])


    next_day_tconfirmed_case_count_mean_max_based=tcr.mean()[0]*max(tc["Total Confirmed"])


    next_day_tconfirmed_case_count_max_mean_based=max(tcr["Total_Confirmed_Ratio"])*tc.mean()[0]


    next_day_tconfirmed_case_count_mean_last_based=tcr.mean()[0]*tc["Total Confirmed"][len(tc)-1]


    next_day_tconfirmed_case_count_last_mean_based=tcr["Total_Confirmed_Ratio"][len(tcr)-1]*tc.mean()[0]


    next_day_tconfirmed_case_count_last_based=tcr["Total_Confirmed_Ratio"][len(tcr)-1]*tc["Total Confirmed"][len(tc)-1]


    next_day_tconfirmed_case_count_middle_based=tcr["Total_Confirmed_Ratio"][3]*tc["Total Confirmed"][3]


    pred_val=[]
    pred_val.append(np.round(next_day_tconfirmed_case_count_mean_based))
    pred_val.append(np.round(next_day_tconfirmed_case_count_max_based))
    pred_val.append(np.round(next_day_tconfirmed_case_count_mean_max_based))
    pred_val.append(np.round(next_day_tconfirmed_case_count_max_mean_based))
    pred_val.append(np.round(next_day_tconfirmed_case_count_mean_last_based))
    pred_val.append(np.round(next_day_tconfirmed_case_count_last_mean_based))
    pred_val.append(np.round(next_day_tconfirmed_case_count_last_based))
    pred_val.append(np.round(next_day_tconfirmed_case_count_middle_based))


    
    if(en<len(data_total_ratio)):
        pred_dict={"date":data_total_ratio["Date"][en-1]+datetime.timedelta(hours=24),"original_val":data_total_ratio["Total Confirmed"][en],"mean_based":pred_val[0],"max_based":pred_val[1],"mean_max_based":pred_val[2],"max_mean_based":pred_val[3],"mean_last_based":pred_val[4],"last_mean_based":pred_val[5],"last_based":pred_val[6],"middle_based":pred_val[7]}
    else:
        pred_dict={"date":data_total_ratio["Date"][en-1]+datetime.timedelta(hours=24),"original_val":0,"mean_based":pred_val[0],"max_based":pred_val[1],"mean_max_based":pred_val[2],"max_mean_based":pred_val[3],"mean_last_based":pred_val[4],"last_mean_based":pred_val[5],"last_based":pred_val[6],"middle_based":pred_val[7]}
        
        
        
    
    
    pred_df=pd.DataFrame(pred_dict, index=[0])
    
    
    
d=pd.read_excel(r"C:\Users\Mark 1\tamprd_df_strd.xlsx")
d=pd.concat([d,pred_df])
d.reset_index(drop=True,inplace=True)
d['original_val'][len(d)-2]=data_total_ratio['Total Confirmed'][len(data_total_ratio)-1]
d.to_excel("tamprd_df_strd.xlsx")
```

```python
err_dict={"date":[],"mipe":[],"mxne":[],"mipv":[],"mxnv":[],"original_val":[],"pprd_val":[],"nprd_val":[]}
psc=[]
ngc=[]
date=[]
ps=[]
ng=[]
org_val=[]
pprd_val=[]
nprd_val=[]
for j,k in zip(range(len(edf)),range(len(tce))):
    
    a=list(edf.loc[j])
    pos=[]
    neg=[]
    for i in a:
        if(i>=0):
            pos.append(i)
        else:
            neg.append(i)
    if(len(pos)):
        aa=min(pos)
    
        psi=a.index(aa)
    
        ps.append(aa)
    
        psc.append(list(edf.columns)[psi])
    
    
        pprd_val.append(tce["original_val"][k]-aa)
    else:
        ps.append("No")
    
        psc.append("No")
    
    
        pprd_val.append("No")
        
    
    if(len(neg)):
        bb=max(neg)
        ngi=a.index(bb)
        ng.append(bb)
        ngc.append(list(edf.columns)[ngi])
        nprd_val.append(tce["original_val"][k]-bb)
    else:
        ng.append("No")
    
        ngc.append("No")
    
    
        nprd_val.append("No")
    
    date.append(str(tce["date"][k]).strip("00:00:00"))
    org_val.append(tce["original_val"][k])
err_dict.update({"date":date})
err_dict.update({"mipe":psc})
err_dict.update({"mxne":ngc})
err_dict.update({"mipv":ps})
err_dict.update({"mxnv":ng})
err_dict.update({"original_val":org_val})
err_dict.update({"pprd_val":pprd_val})
err_dict.update({"nprd_val":nprd_val})


```
