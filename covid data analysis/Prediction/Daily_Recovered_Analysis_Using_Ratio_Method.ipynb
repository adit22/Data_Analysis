```python
data_daily_ratio=data_daily.copy()
```

```python
new_cols=["Daily_Confirmed_Ratio","Total_Confirmed_Ratio","Daily_Recovered_Ratio","Total_Recovered_Ratio","Daily_Deceased_Ratio","Total_Deceased_Ratio"]
existing_cols=list(data_daily_ratio.columns)
existing_cols.pop(0)
existing_cols
for i,j in zip(new_cols,existing_cols):
    data_daily_ratio[i]=list(range(len(data_daily_ratio)))
    data_daily_ratio[i][0]=1
    tmp=[]
    tmp.append(data_daily_ratio[i][0])
    for k in range(0,len(data_daily_ratio)-1):
        tmp.append(data_daily_ratio[j][k+1]/data_daily_ratio[j][k])
    data_daily_ratio[i]=tmp
data_daily_ratio.head()
        
```

```python
strt_cnt=1
end_cnt=8
strt_lst=[]
end_lst=[]
strt_lst.append(strt_cnt)
end_lst.append(end_cnt)
while(end_cnt<=len(data_daily_ratio)):
        strt_cnt+=1
        strt_lst.append(strt_cnt)
        end_cnt+=1
        end_lst.append(end_cnt)
if(end_lst[len(end_lst)-1]>=len(data_daily_ratio)):
    end_lst.pop(len(end_lst)-1)
strt_lst.pop(len(strt_lst)-1)
#print(strt_lst)
#print(end_lst)
#print(len(strt_lst)==len(end_lst))

for st,en in zip(strt_lst,end_lst):

    dcr=data_daily_ratio.iloc[st:en,9:10]
    dcr.reset_index(inplace = True, drop = True)
    dc=data_daily_ratio.iloc[st:en,3:4]
    dc.reset_index(inplace = True, drop = True)



    next_day_recovered_case_count_mean_based=dcr.mean()[0]*dc.mean()[0]

    next_day_recovered_case_count_max_based=max(dcr["Daily_Recovered_Ratio"])*max(dc["Daily Recovered"])


    next_day_recovered_case_count_mean_max_based=dcr.mean()[0]*max(dc["Daily Recovered"])


    next_day_recovered_case_count_max_mean_based=max(dcr["Daily_Recovered_Ratio"])*dc.mean()[0]


    next_day_recovered_case_count_mean_last_based=dcr.mean()[0]*dc["Daily Recovered"][len(dc)-1]


    next_day_recovered_case_count_last_mean_based=dcr["Daily_Recovered_Ratio"][len(dcr)-1]*dc.mean()[0]


    next_day_recovered_case_count_last_based=dcr["Daily_Recovered_Ratio"][len(dcr)-1]*dc["Daily Recovered"][len(dc)-1]


    next_day_recovered_case_count_middle_based=dcr["Daily_Recovered_Ratio"][3]*dc["Daily Recovered"][3]


    pred_val=[]
    pred_val.append(np.round(next_day_recovered_case_count_mean_based))
    pred_val.append(np.round(next_day_recovered_case_count_max_based))
    pred_val.append(np.round(next_day_recovered_case_count_mean_max_based))
    pred_val.append(np.round(next_day_recovered_case_count_max_mean_based))
    pred_val.append(np.round(next_day_recovered_case_count_mean_last_based))
    pred_val.append(np.round(next_day_recovered_case_count_last_mean_based))
    pred_val.append(np.round(next_day_recovered_case_count_last_based))
    pred_val.append(np.round(next_day_recovered_case_count_middle_based))

    if(en<len(data_daily_ratio)):
        pred_dict={"date":data_daily_ratio["Date"][en-1]+datetime.timedelta(hours=24),"original_val":data_daily_ratio["Daily Recovered"][en],"mean_based":pred_val[0],"max_based":pred_val[1],"mean_max_based":pred_val[2],"max_mean_based":pred_val[3],"mean_last_based":pred_val[4],"last_mean_based":pred_val[5],"last_based":pred_val[6],"middle_based":pred_val[7]}
    else:
        pred_dict={"date":data_daily_ratio["Date"][en-1]+datetime.timedelta(hours=24),"original_val":0,"mean_based":pred_val[0],"max_based":pred_val[1],"mean_max_based":pred_val[2],"max_mean_based":pred_val[3],"mean_last_based":pred_val[4],"last_mean_based":pred_val[5],"last_based":pred_val[6],"middle_based":pred_val[7]}
        
    
    pred_df=pd.DataFrame(pred_dict, index=[0])
    
    
    d=pd.read_excel(r"C:\Users\Mark 1\ramprd_df_strd.xlsx")
    d=pd.concat([d,pred_df])
    d.reset_index(drop=True,inplace=True)
    #d['original_val'][len(d)-2]=data_daily_ratio['Daily Recovered'][len(data_daily_ratio)-1]
    d.to_excel("ramprd_df_strd.xlsx")
    
```

```python
err_dict={"date":[],"mipe":[],"mxne":[],"mipv":[],"mxnv":[],"original_val":[],"pprd_val":[],"nprd_val":[]}
psc=[]
ngc=[]
date=[]
ps=[]
ng=[]
org_val=[]
pprd_val=[]
nprd_val=[]
for j,k in zip(range(len(edf)),range(len(dce))):
    
    a=list(edf.loc[j])
    pos=[]
    neg=[]
    for i in a:
        if(i>=0):
            pos.append(i)
        else:
            neg.append(i)
    if(len(pos)):
        aa=min(pos)
    
        psi=a.index(aa)
    
        ps.append(aa)
    
        psc.append(list(edf.columns)[psi])
    
    
        pprd_val.append(dce["original_val"][k]-aa)
    else:
        ps.append("No")
    
        psc.append("No")
    
    
        pprd_val.append("No")
        
    
    if(len(neg)):
        bb=max(neg)
        ngi=a.index(bb)
        ng.append(bb)
        ngc.append(list(edf.columns)[ngi])
        nprd_val.append(dce["original_val"][k]-bb)
    else:
        ng.append("No")
    
        ngc.append("No")
    
    
        nprd_val.append("No")
    
    date.append(str(dce["date"][k]).strip("00:00:00"))
    org_val.append(dce["original_val"][k])
err_dict.update({"date":date})
err_dict.update({"mipe":psc})
err_dict.update({"mxne":ngc})
err_dict.update({"mipv":ps})
err_dict.update({"mxnv":ng})
err_dict.update({"original_val":org_val})
err_dict.update({"pprd_val":pprd_val})
err_dict.update({"nprd_val":nprd_val})


```
