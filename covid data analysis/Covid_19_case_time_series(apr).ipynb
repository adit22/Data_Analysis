```python
dataapr_diff=dataapr.copy()
dataapr_diff.head()
```

```python
new_cols=["Daily_Confirmed_Diff","Total_Confirmed_Diff","Daily_Recovered_Diff","Total_Recovered_Diff","Daily_Deceased_Diff","Total_Deceased_diff"]
existing_cols=list(dataapr_diff.columns)
existing_cols.pop(0)
existing_cols
prv_vals=[309,1635,13,150,6,47]
for i,j,l in zip(new_cols,existing_cols,prv_vals):
    dataapr_diff[i]=list(range(len(dataapr_diff)))
    dataapr_diff[i][0]=dataapr_diff[j][0]-l
    for k in range(0,len(dataapr_diff)-1):
        dataapr_diff[i][k+1]=dataapr_diff[j][k+1]-dataapr_diff[j][k]
dataapr_diff.head()        
```

```python
kys=["daily_cnf_diff_pos_neg","total_cnf_diff_pos_neg","daily_rcvd_diff_pos_neg","total_rcvd_diff_pos_neg","daily_dcsd_diff_pos_neg","total_dcsd_diff_pos_neg"]
new_cols=["Daily_Confirmed_Diff","Total_Confirmed_Diff","Daily_Recovered_Diff","Total_Recovered_Diff","Daily_Deceased_Diff","Total_Deceased_diff"]
pos_neg_dict={}
pos_neg_len_dict={}
a=[]
b=[]
for i,j in zip(kys,new_cols):
    pos=dataapr_diff[dataapr_diff[j]>0]
    a.append(pos)
    b.append(len(pos))
    neg=dataapr_diff[dataapr_diff[j]<0]
    a.append(neg)
    b.append(len(neg))
    zro=dataapr_diff[dataapr_diff[j]==0]
    a.append(zro)
    b.append(len(zro))
    pos_neg_dict.update({i:a})
    pos_neg_len_dict.update({i:b})
    a=[]
    b=[]
```

```python
dataapr_ratio=dataapr.copy()
```

```python
new_cols=["Daily_Confirmed_Ratio","Total_Confirmed_Ratio","Daily_Recovered_Ratio","Total_Recovered_Ratio","Daily_Deceased_Ratio","Total_Deceased_Ratio"]
existing_cols=list(dataapr_ratio.columns)
existing_cols.pop(0)
existing_cols
prv_vals=[309,1635,13,150,6,47]
for i,j,l in zip(new_cols,existing_cols,prv_vals):
    dataapr_ratio[i]=list(range(len(dataapr_ratio)))
    dataapr_ratio[i][0]=round(dataapr_ratio[j][0]/l,2)
    tmp=[]
    tmp.append(dataapr_ratio[i][0])
    for k in range(0,len(dataapr_ratio)-1):
        tmp.append(round(dataapr_ratio[j][k+1]/dataapr_ratio[j][k],2))
    dataapr_ratio[i]=tmp
dataapr_ratio.head()
        
```
