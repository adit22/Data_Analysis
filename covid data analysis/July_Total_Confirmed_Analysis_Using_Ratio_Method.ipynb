```python
err_dict={"date":[],"mxpe":[],"mnne":[],"mipv":[],"mxnv":[]}
psc=[]
ngc=[]
date=[]
ps=[]
ng=[]
for j,k in zip(range(len(edf)),range(len(tc)-1)):
    
    a=list(edf.loc[j])
    pos=[]
    neg=[]
    for i in a:
        if(i>=0):
            pos.append(i)
        else:
            neg.append(i)
    aa=min(pos)
    bb=max(neg)
    psi=a.index(aa)
    ngi=a.index(bb)
    ps.append(aa)
    ng.append(bb)
    psc.append(list(edf.columns)[psi])
    ngc.append(list(edf.columns)[ngi])
    date.append(str(tc["date"][k]).strip("00:00:00"))
    
err_dict.update({"date":date})
err_dict.update({"mxpe":psc})
err_dict.update({"mnne":ngc})
err_dict.update({"mipv":ps})
err_dict.update({"mxnv":ng})


```
